```python

#Steps to follow:

1. Note down the null and alternate hypothesis
2. Create a new or sub dataset based on the condition (Problem Statement)
3. Now the new data will be our entire population, from this population we will create a new sample
4. We will perform hypothesis testing and do the conclusion
```

```python
#Q1. Lets suppose it was ovserved that the average purchase made by the Men of age group 18-25 was 10000. Is it still the same


H0: Avg purchase made by the Men of age group 18-25 is 10000 (no sign difference)
Ha: Avg purchase made by the Men of age group 18-25 is not 10000 (sign difference)
```

```python
new_data = df[(df['Age']==1) & (df['Gender']==1)]  	#age ==1 (18-25 group & Gender == 1(Male))
new_data
```

```python
sample_size = 1000  	#we select 10-20% data as sample

new_data_sample = new_data.sample(sample_size, random_state = 20)	#random_state -> it will behave like seed
new_data_sample
```

```python
#one sample ttest

from scipy.stats import ttest_1samp

t_stat, p_value = ttest_1samp(new_data_sample['Purchase'], 10000)     #10000 value we are checking (given in the ques)
```

```python
p_value
```

```python
if p_value < 0.05:
	print ('Reject H0: Avg purchase made by the Men of age group 18-25 is not 10000 (sign difference)' )

else:
	print('Accept H0: Avg purchase made by the Men of age group 18-25 is 10000 (no sign difference)')
```

```python
#Q2. Is the average purhcase made by men and women of age 18-25 is same?

H0 = average purhcase made by men and women of age 18-25 is same
Ha = average purhcase made by men and women of age 18-25 is not same
```

```python
#we need to select our populatopm
- first group (men)-> 18-25
- second group (women)-> 18-25
```

```python
#first group (men)-> 18-25
new_data_male = df[(df['Age']==1) & (df['Gender']==1)]            #All male (Gender == 1) of age 18-25 group (age == 1)
new_data_male
```

```python
sample_size = 600

new_sample_male = new_data_male.sample(sample_size, random_state = 0)
new_sample_male
```

```python
#second group (women)-> 18-25
new_data_female = df[(df['Age']==1) & (df['Gender']==0)]        #All female (Gender == 0) of age 18-25 group (age == 1)
new_data_female
```

```python
sample_size = 600

new_sample_female = new_data_female.sample(sample_size, random_state = 0)
new_sample_female
```

```python
from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(new_sample_male['Purchase'], new_sample_female['Purchase'])
```

```python
p_value
```

```python
if p_value < 0.05:
	print ('Reject H0: average purhcase made by men and women of age 18-25 is not same' )

else:
	print('Accept H0: average purhcase made by men and women of age 18-25 is same')
```

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
total_count = [df[(df['Gender']== 1) & (df['Age'] == 1)].shape[0], df[(df['Gender']== 1) & (df['Age'] == 2)].shape[0]]
total_count
```

```python
t_stats, p_value = proportions_ztest(count = [1000,1000], nobs = total_count)         #sample_size = [900,900]  and nobs = no of observation
```

```python
if p_value < 0.05:
	print ('Reject H0: Men who have spend more than 10000 are not same for the age 18-25 and 26-35' )

else:
	print('Accept H0: Men who have spend more than 10000 are same for the age 18-25 and 26-35')
```
