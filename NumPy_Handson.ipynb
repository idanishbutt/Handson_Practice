```python
#9. Given a numpy array that has values = [1,2,3,4,5]. Perform the following operations.
a. Insert the element 6 at the end of the array.
b. Insert the element 10 at the beginning of the array.
c. Insert the element 30, before 4.
d. Update the element 5 as 50.
e. Delete the element 50 in the updated array.
f. Create a function to search for the element 30.
g. Sort the array in ascending order
```

```python
import numpy as np
arr= np.array([1,2,3,4,5])
arr
```

```python
#a. Insert the element 6 at the end of the array.
arr1 = np.append(arr,6)
arr1
```

```python
#b. Insert the element 10 at the beginning of the array.
arr2 = np.insert(arr,0,10)            #(array, index, value)
arr2
```

```python
#c. Insert the element 30, before 4.
arr = np.array([1, 2, 3, 4, 5])
index_4 = np.where(arr == 4)[0][0]
arr = np.insert(arr, index_4, 30)
arr

# arr == 4: Checks each element and returns True for the ones equal to 4.
# np.where(arr == 4): Finds the index (or indices) where that condition is True.
# [0][0]: This extracts the first matching index. (Just in case there are multiple 4s.)
```

```python
#d. Update the element 5 as 50.
arr = np.array([1, 2, 3, 4, 5])
index_of_5 = np.where(arr == 5)[0][0]         #arr[arr == 5] = 50
arr[index_of_5] = 50

arr
```

```python
#e. Delete the element 50 in the updated array.
arr = np.array([1, 2, 3, 4, 50])
index_of_50 = np.where(arr == 50)[0][0]       #arr = arr[arr != 50]
arr = np.delete(arr, index_of_50)

arr
```

```python
#f. Create a function to search for the element 30.
def search_element(array, value):
    if value in array:
        print(f"{value} found at index:", np.where(array == value)[0][0])
    else:
        print(f"{value} not found in the array.")

search_element(arr, 30)
```

```python
#g. Sort the array in ascending order
a = np.array([2,6,41,20,33,20])
arr = np.sort(a)

arr
```
