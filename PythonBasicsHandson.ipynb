```python
#Armstrong Number

def armstrong_number():
  num=int(input("enter a number: "))    #we have done typecasting here because there is only integer number is used (by default it is string which accepts everything)
  power=len(str(num))                   #we have done typecasting again just to calcalte length fxn

  sum=0
  for i in str(num):
    sum+=int(i)**power
  if sum==num:
    print(f"{num} is armstrong number")
  else:
    print(f"{num} is not armstrong number")
  armstrong_number()
```

```python
#prime number
def prime_number():
  num=int(input(":enter a number: "))

  if num>=1:
    for i in range(2,num):
      if num%i==0:
        print(f"{num} is not a prime")
        break
    else:
      print(f"{num} is a prime")
prime_number()
```

```python
#7. Create a calculator program for addition, subtraction, multiplication and floor division. Take the inputs from the user, and based on the choice of operation, return the results.

def calculator():
  num1= int(input("enter a number: "))
  num2= int(input("enter a number: "))

  select_operation= input("Choose operator (+, -, *, //)")

  if select_operation== "+":
    result= num1 + num2
  elif select_operation== "-":
    result= num1 - num2
  elif select_operation== "*":
    result= num1 * num2
  elif select_operation== "//":
    if num2!=0:
      result= num1 // num2
    else:
      return "Error: Division by zero"
  else:
    return "Invalid Operation"

  return f"Result: {result}"
calculator()
```
