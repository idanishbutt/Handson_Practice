```python
#1. Create scatter plots for the following data, make sure all the appear in the same plane with x-label and and y-labels evenly placed

- Create scatter plots with between (mpg-hp, mpg-weight), and (hp-mpg,hp-time-to-60) with respect to each brand
```

```python
fig = plt.subplots(2,2, figsize =(20,10))

#mpg-hp
plt.subplot(2,2,1)  #focuses on the first graph
sns.scatterplot(x= df['mpg'], y =df['hp'], hue=df['brand'])       #hue - differentiates categories on the plot wrt a column
plt.title ('MPG vs HP')
plt.xlabel('MPG')
plt.ylabel('HP')

#mpg-weight
plt.subplot(2,2,2)  #focuses on the second graph
sns.scatterplot(x= df['mpg'], y =df['weightlbs'], hue=df['brand'])
plt.title ('MPG vs WEIGHT')
plt.xlabel('MPG')
plt.ylabel('WEIGHT')

#hp-mpg
plt.subplot(2,2,3)  #focuses on the third graph
sns.scatterplot(x= df['hp'], y =df['hp'], hue=df['brand'])
plt.title ('HP vs MPG')
plt.xlabel('HP')
plt.ylabel('MPG')

#hp-time-to-60
plt.subplot(2,2,4)  #focuses on the fourth graph
sns.scatterplot(x= df['hp'], y =df['time-to-60'], hue=df['brand'])
plt.title ('HP vs TIME-TO-60')
plt.xlabel('HP')
plt.ylabel('TIME-TO-60')


plt.show()
```

```python
fig = plt.subplots(1,1, figsize =(10,50))

plt.subplot(1,1,1)  #focuses on the second graph
sns.scatterplot(x= df['mpg'], y =df['weightlbs'], hue=df['brand'])
plt.title ('MPG vs WEIGHT')
plt.xlabel('MPG')
plt.ylabel('WEIGHT')
```

```python
#2. Create bar plots for the following, make sure all the plots appear in the same plane with x- labels and y-labels evenly placed

- Create a bar plot that shows the visual representation of hp, mpg, weight, time-to-60 with respect to the number of cylinders for each of the brands.
```

```python
figure = plt.subplots(2,2, figsize = (20,10))

#HP vs Cylinder
plt.subplot(2,2,1)
sns.barplot(x = df['hp'], y = df['cylinders'], hue = df['brand'])
plt.title('HP vs Cylinder')
plt.xlabel('HP')
plt.ylabel('Cylinder')

#MPG vs Cylinder
plt.subplot(2,2,2)
sns.barplot(x = df['mpg'], y = df['cylinders'], hue = df['brand'])
plt.title('MPG vs Cylinder')
plt.xlabel('MPG')
plt.ylabel('Cylinder')

#WEIGHT vs Cylinder
plt.subplot(2,2,3)
sns.barplot(x = df['weightlbs'], y = df['cylinders'], hue = df['brand'])
plt.title('WEIGHT vs Cylinder')
plt.xlabel('WEIGHT')
plt.ylabel('Cylinder')

#HP vs Cylinder
plt.subplot(2,2,4)
sns.barplot(x = df['time-to-60'], y = df['cylinders'], hue = df['brand'])
plt.title('HP vs Cylinder')
plt.xlabel('Time-to-60')
plt.ylabel('Cylinder')

plt.show()
```
